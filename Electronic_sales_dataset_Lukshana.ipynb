```python
# importing relevant libraries for K means clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# selecting age and total price
df = electronic_sales[['Age', 'Total Price']].copy()
X = df[['Age', 'Total Price']]

# scaling the age and total price
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# elbow method
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init="k-means++", random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(12, 6))
plt.plot(range(1, 11), wcss, linewidth=2, color="orange", marker="8")
plt.grid()
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal K")
plt.show()

# applying KMeans with chosen K
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# visualising the clusters
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['Cluster'], cmap='rainbow', alpha=0.6)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', s=200)
plt.xlabel('Age (scaled)')
plt.ylabel('Total Price (scaled)')
plt.title('Clusters of Customers')
plt.legend()
plt.grid()
plt.show()

# inverse transforming cluster centers
centroids_original = scaler.inverse_transform(kmeans.cluster_centers_)
centroids_df = pd.DataFrame(centroids_original, columns=['Age', 'Total Price'])
print("\nCentroids in Original Scale:\n", centroids_df)

# printing the cluster sizes
print("\nCluster sizes:\n", df['Cluster'].value_counts())
```
